```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3))
    return tracks
```

```python
def get_features(uri):
    features =[]
    count = 0
    for x in uri:
        count = count+1
        if count %15 == 0:
            sleep(randint(1,2))
        try:
            features.append(sp.audio_features(x)) 
        except:
            features.append(None)
    return features
```

```python
data= pd.DataFrame({"Artists":Artists,"Title":Title, "Song_uri": Song_uri, "audio_features":audio_features})
```

```python
data
```

```python
def flattendf(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
columns = ['audio_features']
music= flattendf(data, columns)
music.head(5)
```

```python
pd.set_option('display.max_columns', 0)
columns = ['audio_features_0']
all_songs = flattendf(music, columns)
all_songs
```
